```python
import pandas as pd
```

```python
df = pd.read_csv(
    "../Data/raw_ai_student_impact_dataset (1).csv"
)

df.head()
```

```python
df.info()

df.isnull().sum()

df.duplicated().sum()
```

```python
for col in ["Major_Category", "Year_of_Study", "Primary_Use_Case", "Burnout_Risk_Level"]:
    print("\n", col)
    print(df[col].value_counts())
```

```python
df["gpa_change"] = (
    df["Post_Semester_GPA"]
    - df["Pre_Semester_GPA"]
)
df.head()
```

```python
# GPA CHANGE PER MAJOR
df.groupby(
    "Major_Category"
)["gpa_change"].mean()
```

```python
#* GPA CHANGE PER SUBCRIPTION ( FREE OR PAID )
df.groupby(
    "Paid_Subscription"
)["gpa_change"].mean()
```

```python
# Average Gpa change of Students who used ai grouped by how many hours do they use ai at a weekly basis 
df.groupby(
    pd.cut(df["Weekly_GenAI_Hours"], bins=5)
)["gpa_change"].agg(["mean", "count"])
```

```python
# Retention rate of students based on their use of ai weekly
df.groupby(
    pd.cut(df["Weekly_GenAI_Hours"], bins=5)
)["Skill_Retention_Score"].agg(["mean"],)
```

```python
df.groupby(
    pd.cut(df["Anxiety_Level_During_Exams"], bins = 10)
)["Weekly_GenAI_Hours"].agg(["mean", "count"])
```

```python
df.to_csv(
    "../Data/cleaned_ai_student.csv",
    index=False
)
```
